```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in csv file
df = pd.read_csv(r'C:\Users\johnc\Desktop\movies.csv')
```

```python
# View data

df
```

```python
# Identify data types

df.dtypes
```

```python
# Clear duplicates

df.drop_duplicates()
```

```python
# Sort data by gross

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Sort data by budget

df.sort_values(by = ['budget'], inplace = False, ascending = False)
```

```python
# Sort data by votes

df.sort_values(by = ['votes'], inplace = False, ascending = False)
```

```python
# Check which, if any, data points are strongly correlated

# Pearson = parametric correlation

df.corr(method = 'pearson')
```

```python
# Represent Pearson Correlation with matrix

# Matrix indicates strongest correlation between budget and gross

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Pearson Correlation Matrix")

plt.show()
```

```python
# Plot budget vs. gross using regression plot

sns.regplot(x = "budget", y = "gross", data = df)
```

```python
# Check which, if any, data points are strongly correlated

# Kendall = non-parametric correlation

df.corr(method = 'kendall')
```

```python
# Represent Kendall Correlation with matrix

# Matrix indicates strongest correlation between votes and gross

correlation_matrix = df.corr(method = 'kendall')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Kendall Correlation Matrix")

plt.show()
```

```python
# Plot votes vs. gross using regression plot

sns.regplot(x = "votes", y = "gross", data = df)
```

```python
# Tally total gross per director

director_gross_sum = df.groupby(['director'])[["gross"]].sum()

# Sort 15 highest grossing directors

director_gross_sum_sorted = director_gross_sum.sort_values(['gross','director'], ascending = False)[:15]

# Cast total as 64 bit integer

director_gross_sum_sorted = director_gross_sum_sorted['gross'].astype('int64') 

# Output

director_gross_sum_sorted
```

```python
# Tally total gross per company

company_gross_sum = df.groupby(['company'])[["gross"]].sum()

# Sort 15 highest grossing companies

company_gross_sum_sorted = company_gross_sum.sort_values(['gross','company'], ascending = False)[:15]

# Cast total as 64 bit integer

company_gross_sum_sorted = company_gross_sum_sorted['gross'].astype('int64') 

# Output

company_gross_sum_sorted
```
